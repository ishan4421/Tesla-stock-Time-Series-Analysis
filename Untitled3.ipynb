```python
tesla= pd.read_csv('Tesla Dataset.csv')
```

```python
tesla
```

```python
tesla['Date']= pd.to_datetime(tesla['Date'])
```

```python
tesla.set_index('Date', inplace=True)
```

```python
tesla
```

```python
plt.plot(tesla['Close'])
plt.show()
```

```python
#now will do some analysis on this
```

```python
#time series data has 3 main components: seasonality, trends, residuals
```

```python
#before going forward, will try to look at somme autocorrelation function to find the cyclic period for decomposition
```

```python
plot_acf(tesla['Close'], lags=90)
plt.show()
```

```python
decompose= seasonal_decompose(tesla['Close'], model='multiplicative', period=90 )
```

```python
plt.figure(figsize=(10,4))
plt.plot(decompose.trend)
plt.title('Trend')
plt.show()

plt.figure(figsize=(16,3))
plt.plot(decompose.seasonal)
plt.title('Seasonal')
plt.show()

plt.figure(figsize=(10,6))
plt.plot(decompose.resid)
plt.title('Residual')
plt.show()
```

```python
#now plotting acf on trend, seasonal and residuals
plot_acf(decompose.trend, lags=50)
plt.show()

plot_acf(decompose.seasonal, lags=90)
plt.show()

plot_acf(decompose.resid, lags=90)
plt.show()
```
